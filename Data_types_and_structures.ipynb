2. Explain the difference between mutable and immutable data types with
    examples.
   - Data types in Python are categorized into mutable and immutable data types. Mutable data types are those whose values can be changed, whereas immutable data types are ones in which the values can’t be changed.
   Mutable Data Type in Python
In Python, a data type is mutable if its values can be changed, updated, or modified after the data type has been created. In other words, once an object of the mutable data type is initialized, you can update its content without creating a new object.
Examples of Mutable Data Type:
Python Lists
Python lists methods are mutable; the list’s contents are modified in place. You can change a list in place, add new elements, overwrite existing elements, or delete them. Also, note that the same value can occur in a list more than once.
Immutable Data Type in Python
Data types in Python whose values can’t be changed or modified are known as Immutable Data Types. Once the value or the object of the immutable data type is initialized, its value remains constant throughout its lifetime.

But, if you want to change the original object, a new object will be created with the modified values.

Common Immutable Data Types in Python are:

Integers (int)
It represents whole numbers, both positive and negative.
Once an integer object is created, its value remains fixed. You cannot change the value of an existing integer object.

5.Why might you use a set instead of a list in Python?
  - A set in Python is an unordered collection of unique elements, while a list is an ordered collection that can contain duplicates. Sets and lists are used for different purposes, and the choice between them depends on the specific requirements of the task.
Here's why you might use a set instead of a list:
Uniqueness:
If you need to ensure that all elements in a collection are unique, a set is the ideal choice. Sets automatically discard duplicate values, which can be useful for tasks like removing duplicates from a list or checking for membership.
Membership testing:
Sets are highly optimized for membership testing. Checking if an element exists in a set is much faster than checking if it exists in a list, especially for large collections.
Set operations:
Sets provide efficient ways to perform mathematical set operations like union, intersection, and difference. These operations can be useful for tasks like comparing collections or finding common elements.
Performance:
Sets generally have better performance than lists for tasks like membership testing and removing duplicates. This is because sets are implemented using hash tables, which allow for fast lookups.

8.What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table is a data structure that stores key-value pairs, where each key is unique and its location is determined by a hash function. This function takes the key as input and generates an index, which is then used to store the associated value in an array-like structure. Hash tables allow for fast average-case time complexity, O(1), for insertion, deletion, and search operations.
In Python, dictionaries are implemented using hash tables. When a key-value pair is added to a dictionary, Python calculates the hash value of the key and uses it to determine the storage location for the value. This mechanism enables efficient retrieval of values based on their keys.
While hash tables provide excellent average-case performance, collisions (when different keys produce the same hash value) can occur. Various collision resolution techniques, like separate chaining or open addressing, are employed to handle these situations.

10.Explain why strings are immutable in Python?
   - Strings in Python are immutable because of design choices that prioritize performance, security, and simplicity. Immutability ensures that a string's value remains constant throughout its existence. When a string is modified, instead of changing the original string in memory, a new string object is created with the updated value.
One key reason for immutability is performance optimization. Since strings are used extensively, Python can implement optimizations like string interning, where identical string literals are stored only once in memory. This optimization wouldn't be possible if strings were mutable, as changing one instance would affect all others.
Immutability also enhances security. When a string is passed to a function or used in a data structure, it's guaranteed that its value won't change unexpectedly. This prevents unintended side effects and makes code more predictable and easier to debug.
Furthermore, immutability simplifies code reasoning. Developers can rely on the fact that a string's value will remain constant, making it easier to understand and maintain code. It also enables the use of strings as dictionary keys, which wouldn't be possible if they were mutable.

17. Describe the time complexity of accessing elements in a dictionary?
    - Accessing elements in a dictionary (or hash map) typically has an average time complexity of O(1), meaning it takes constant time regardless of the dictionary's size. However, in the worst-case scenario (e.g., many collisions), it can degrade to O(n). Dictionaries and Hash Tables:
Dictionaries are often implemented using hash tables, which are data structures that allow for fast lookups based on keys.
Hash Function:
A hash function maps keys to indices within the hash table (or dictionary).
Average Case (O(1)):
On average, accessing an element in a dictionary involves calculating the hash of the key, which takes constant time, and then accessing the corresponding element in the hash table, which also takes constant time.
Worst Case (O(n)):
In the worst-case scenario, multiple keys might hash to the same index (a collision), leading to a linear search within the "bucket" (or linked list) at that index, resulting in O(n) time complexity.

20.Explain the difference between a list and a dictionary in terms of data retrieval.
   - In terms of data retrieval, lists use numerical indices (starting from 0) for accessing elements, while dictionaries use keys to access associated values, offering significantly faster lookups than lists.
Here's a more detailed breakdown:
Lists:
Access: Elements in a list are accessed by their position (index), which is a numerical value starting from 0.
Retrieval Speed: Searching for a specific element in a list can be slow, especially for large lists, as it requires iterating through the list until the element is found (linear time complexity, O(n)).
Example: my_list = [10, 20, 30, 40] To access the value 30, you would use my_list[2].
Dictionaries:
Access: Elements in a dictionary are accessed by their unique key, which can be any immutable data type (like strings, numbers, or tuples).
Retrieval Speed: Dictionaries use a hash table, allowing for very fast lookups (average-case constant time complexity, O(1)).
Example: my_dict = {"name": "Alice", "age": 30} To access the value 30, you would use my_dict["age"].
